```python
# import necessary liabrary
import numpy as np
import pandas as pd
```

```python
# Load Data
data = pd.read_csv('Sales data.csv')
pd.set_option('display.max_columns',None)
```

```python
data.head()
```

```python
data.columns
```

## Q 1. What is the total sales value for each product category and sub-category?

```python
product_sales_pivot = pd.pivot_table(data,index=['Category','Sub-Category'],values='Sales',aggfunc='sum')
product_sales_pivot
```

## Q2. What is the minimum, maximum, and average sales value for each product category?

```python
sales_pivot = pd.pivot_table(data,index=['Category'],values='Sales',aggfunc={'min','max','mean'})
sales_pivot
```

## Q 3. What is the minimum, maximum, and average profit generated by each customer segment?

```python
pivot_df = data.pivot_table(index='Segment',values='Profit',aggfunc={'min','max','mean'})
pivot_df
```

## Q 4.How does the minimum, maximum, and average quantity sold vary across different product categories in each region?

```python
pivot_df1 = data.pivot_table(index=['Region','Category'], values='Quantity', aggfunc={'min','max','mean'})
pivot_df1
```

## 5. Which ship mode has the highest average profit margin?


```python
profit_margin = data.pivot_table(index='Ship Mode',values='Profit',aggfunc='mean')
profit_margin['Profit'].sort_values(ascending=False)
```

## 6. How does the profit vary across different customer segments?


```python
cust_segment_profit = data.pivot_table(index='Segment',values='Profit',aggfunc={'min','max','mean','median'})
cust_segment_profit
```

## 7. What is the total quantity sold for each product category in each region?


```python
df2 = data.pivot_table(index=['Region','Category'], values='Quantity', aggfunc='sum')
df2
```

## 8. Which customer has placed the most orders, and what is the total sales amount for that customer?


```python
df3 = data.pivot_table(index='Customer Name',values='Sales', aggfunc={'sum','count'} )
df3.sort_values(ascending=False, by='sum')
```

## 9. How does the average price differ between different product categories?


```python
df4 = data.pivot_table(index='Category', values='Price', aggfunc='mean')
df4
```
