## 3. Create a DataFrame Employee_data using Pandas to store this information and answer the Following Qusetions.



```python
df1 = pd.DataFrame({"Name":["Amit","Akshay","Ankit","Aditya","Rupesh"],
                    "Department":["HR","Engineering","Sales","HR","Sales"],
                    "Position":["Manager","Developer","Executive","Assistant","Manager"],
                    "Salary":[60000,70000,55000,45000,62000]})
df1
```

```python
# 1. Retrieve the salary of the employee named 'Ankit'.
print(f"Salary of Aniket is: {df1['Salary'][0]}")

# 2. Retrieve the department of the employee in the third row.
print(f"Employee of the department in the third row is:{df1['Department'][3]}")

# 3. Retrieve the position of the employee with a salary of 62000.
print(f"Position of the employee with salary 62000: {df1[df1['Salary'] == 62000]['Position'].values[0]}")


# 4. Retrieve the names of all employees whose salary is greater than 50000
print("Employee with salary greater than 50000 is:\n",df1[df1["Salary"] > 50000]['Name'])
```

## 4. Create a DataFrame Employee_data using Pandas to store this information and answer the Following Qusetions. 



```python
df2 = pd.DataFrame({"Product":["Shoes","Bat","T-shirt","Shoes","Bat","T-shirt"],
                    "Region":["North","North","North","South","South","South"],
                    "Sales":[100,150,100,120,120,90],
                    "Profit":[20,30,25,22,28,25]})
df2
```

```python
# 1. Retrieve the sales figure of product 'Bat' in the North region.
print("Sales figure of product bat in North region is: ",df2[(df2["Product"] == 'Bat') & (df2["Region"] == 'North')]["Sales"].values[0])

# 2. Retrieve the total sales in the South region.
print("Total Sales in the south region:", df2[df2["Region"] == "South"]["Sales"].sum())

# 3. Retrieve the total profit for product 'T-Shirts'.
print("Total profit for product T-shirt is:",df2[df2["Product"] == 'T-shirt']['Profit'].sum())

# 4. Retrieve the average profit for all products.
print("Average profit of all product:",df2["Profit"].mean())
```
