 book_issue - book_id(FK), member_id(Fk), issue_date, return_date 
 Also write a function that how many days a book is availble to students """

```python
import psycopg2 as ps
from datetime import datetime

class Library:
    def __init__(self):
        # Use a context manager for better resource management
        self.connection = ps.connect(database='insititue', user='postgres', password='Mukesh@4912')
        with self.connection:
            self.cursor = self.connection.cursor()
            print("Database connected successfully......\n")
            print("Cursor is created successfully.......\n")
            print("------------------------------------------")

    def create_tables(self):
        # Creating Book table
        query = 'DROP TABLE IF EXISTS book cascade'
        self.cursor.execute(query)
        query = 'CREATE TABLE book(book_id SERIAL PRIMARY KEY, title VARCHAR(255) NOT NULL, quantity INT NOT NULL)'
        self.cursor.execute(query)
        print("Book table is created.....\n")

        # Creating Member table
        query = 'DROP TABLE IF EXISTS member cascade'
        self.cursor.execute(query)
        query = '''CREATE TABLE member(member_id SERIAL PRIMARY KEY, 
                                       name VARCHAR(255) NOT NULL, 
                                       address VARCHAR(255) NOT NULL, 
                                       phone VARCHAR(14) NOT NULL)'''
        self.cursor.execute(query)
        print("Member table is created.....\n")

        # Creating Author table
        query = 'DROP TABLE IF EXISTS author cascade'
        self.cursor.execute(query)
        query = '''CREATE TABLE author(author_id SERIAL PRIMARY KEY, 
                                       name VARCHAR(60) NOT NULL, 
                                       nationality VARCHAR(60) NOT NULL)'''
        self.cursor.execute(query)
        print("Author table is created.....\n")

        # Creating Book_issue table
        query = '''CREATE TABLE IF NOT EXISTS book_issue(
                       book_id SERIAL, 
                       member_id SERIAL, 
                       issue_date DATE NOT NULL, 
                       return_date DATE NOT NULL,
                       FOREIGN KEY(book_id) REFERENCES book(book_id),
                       FOREIGN KEY(member_id) REFERENCES member(member_id)
                   )'''
        self.cursor.execute(query)
        print("Book issue table is created.....\n")
        print("------------------------------------------")

        self.connection.commit()

    def insert_records(self):
        # Inserting records into Book table
        query = '''INSERT INTO book (title, quantity) VALUES 
                   ('Python for Beginners Book', 10), 
                   ('Human Compatible', 15), 
                   ('Machine Learning', 20), 
                   ('Artificial Intelligence', 25)'''
        self.cursor.execute(query)
        print("Records inserted into Book table.....\n")

        # Inserting records into Member table
        query = '''INSERT INTO member (name, address, phone) VALUES 
                   ('Mukesh', 'Nashik', '+91 8788941673'), 
                   ('Aditya', 'Mumbai', '+91 7219156953'), 
                   ('Rohit', 'Nagpur', '+91 9022322490'), 
                   ('Vaibhav', 'Malegaon', '+91 7745077696')'''
        self.cursor.execute(query)
        print("Records inserted into Member table.....\n")

        # Inserting records into Author table
        query = '''INSERT INTO author (name, nationality) VALUES 
                   ('Max Tegmark', 'Korean'), 
                   ('Stuart Russell', 'German'), 
                   ('P. Murphy', 'American'), 
                   ('Nils J. Nilsson', 'American')'''
        self.cursor.execute(query)
        print("Records inserted into Author table.....\n")

        # Inserting records into Book_issue table
        query = '''INSERT INTO book_issue (book_id, member_id, issue_date, return_date) VALUES 
                                          (1, 1, '2024-03-01', '2024-03-15'), 
                                          (2, 2, '2024-03-02', '2024-03-16'), 
                                          (3, 3, '2024-03-03', '2024-03-17'), 
                                          (4, 4, '2024-03-04', '2024-03-18')'''
        self.cursor.execute(query)
        print("Records inserted into Book_issue table.....\n")

        self.connection.commit()
        
    def calculate_days(self, book_id):
        # Get the return date for the given book_id
        query = f"SELECT return_date FROM book_issue WHERE book_id = {book_id}"
        self.cursor.execute(query)
        return_date = self.cursor.fetchone()[0]

        # Calculate the difference between today's date and the return date
        today_date = datetime.now().date()
        return_date = datetime.strptime(str(return_date), "%Y-%m-%d").date()
        available_days = (return_date - today_date).days

        return available_days

    def close_connection(self):
        # Use a context manager for better resource management
        with self.connection:
            self.cursor.close()
        print("Connection closed......\n")

```

```python
# Object Creation
lb = Library()

# Creat table object
lb.create_tables()

# Create insert records object
lb.insert_records()

```

```python
# Create object of calculate_available_days
BOOK_ID = 3 # Assume book ID to check
days = lb.calculate_days(BOOK_ID)
print(f"Book with ID {BOOK_ID} is available for {days} days.")

# Create object to close the connection
lb.close_connection()
```
